#1. 개요
최근에 가까운 일본이나 우리나라에서도 지진이 일어나면서 지진이 기후변화에 영향을 받는다는 사실을 접하게됐다. 기후변화란 오랜 기간에 걸쳐 나타나는 평균적인 날씨 패턴의 변화을 뜻한다. 기후변화가 지진에도 영향을 주는것이라면 다른부분에도 영향을 줄것이라고 추측을 하였는데 사람들의 건강에도 영향을 미친다는것을 새롭게 알게되었다.

기후변화 정부 간 협의체(IPCC) '6차 기후변화 종합보고서'에서는 2021~2040년 지구 평균 온도가 1.5℃ 이상 상승할 것이라고 경고했다.
WHO의 Tedros Adhanom Ghebreyesus 사무총장은 "긴급한 기후행동(climate action)이 필요한 이유는 미래가 아닌 현재 건강에 영향을 미치기 때문"이라며 "기후위기는 비전염성질환 발생률을 높이고 보건의료 인력과 인프라를 크게 위협할 수 있는 건강위기"라고 피력했다.

또한 미국 하와이대, 위스콘신대 등 공동연구팀의 학술지 네이처 기후변화 논문에서 기후 위험은 무분별하게 사람들의 이동을 통해 병원체와 새로 접촉하도록 만들어 질병과 전염의 경로의 확산을 막을 수 없다고 하였다.



* 기후위험이 감염 유형을 통해 질환을 악화시키는 경로. 선의 굵기는 고유한 감염성 질병의 수를 나타낸다. 색상은 질병과 비례해, 어두운 색상은 많은 야을 나타내고 밝은 색상은 적은 양을 나타낸다. 숫자는 고유한 감염성 질병의 수를 나타낸다. <'네이처 기후변화'제공>*
20220814500654.jpg

이에 따라 질병관리청은 기후영향에 따른 국민건강피해 최소화를 위해 2022년부터 기후보건포럼을 개최했다.

2023년에 '2023년 제2차 기후보건포럼'에서는 '기후변화 적응대책의 건강보호 전략'을 주제로 건강영향평과와 대응대책을 논의하였다.

기후변화에 관한 활동들과 기사들을 통해 사람들의 건강에도 영향을 끼친다는 것의 심각성을 깨달았고
서울시에 관한 데이터를 수집하여 질병 4가지(식중독, 피부암, 알츠하이머, 심장질환)와 기후변화와의 관계성을 따져보고자한다.

##-수집 방법
국민건강보험공단의 경우 데이터를 따로 신청할 수 있고, 연구 DB, 통계자료를 가져올 수 있다.

서울시 폭염일수: 기상청 기상자료개발포털에서 원하는 범주 선택 후 csv 파일 다운로드

서울시 열대야 일수: 기상청 기상자료개발포털에서 원하는 범주 선택 후 csv 파일 다운로드

서울시 상대습도/ 평균온도: 서울 열린 데이터광장에서 원하는 범주 선택 후 csv 파일 다운로드

서울 연도별 식중독 : 식품의약품안전처에서 연도당 월별 식중독 환자 csv파일을 다운로드 후 하나로 합침.

서울시 피부암 등록환자: 국민건강보험공단, 건강보험심사평가원에서 파일을 다운받은 후 파일에서 서울시 이외의 데이터는 제거

서울시 알츠하이머 병 사망자수 및 사망률: 사망원인통계 통계청에서 csv파일 다운로드 후 연령별 데이터는 제거함.

서울시 심장질환 사망자수 및 사망률: 사망원인통계 통계청에서 csv파일 다운로드 후 연령별 데이터는 제거함.



##-수집된 데이터의 구조 및 특성 설명

#추후계획

알아보고, 추후 기후가 어떻게 변화하느냐에 따른 질병의 발병률, 사망
인수를 예측해볼것이다.
